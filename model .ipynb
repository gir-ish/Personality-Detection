```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

data_path = 'C:/Users/giris/Downloads/RESEARCH/Personality/personality_data/Video_Interview_features.csv'
data = pd.read_csv(data_path)

X = data.drop(['file_name', 'Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness'], axis=1)
y = data[['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.1, random_state=42)

# Define the MLP model
model_dnn = Sequential([
    Dense(256, input_shape=(X_train.shape[1],), kernel_initializer='he_normal'),
    Activation('elu'),
    Dense(128, kernel_initializer='he_normal'),
    Activation('elu'),
    Dense(64, kernel_initializer='he_normal'),
    Activation('elu'),
    Dense(y.shape[1], activation='linear')
])

model_dnn.compile(optimizer=Adam(), loss='mse', metrics=['mae'])

history = model_dnn.fit(X_train, y_train, validation_split=0.2, epochs=1, batch_size=128, verbose=1)

test_loss, test_mae = model_dnn.evaluate(X_test, y_test, verbose=0)

y_pred = model_dnn.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred, multioutput='raw_values'))
mae = mean_absolute_error(y_test, y_pred, multioutput='raw_values')

for i, trait in enumerate(['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness']):
    print(f"{trait} - RMSE: {rmse[i]}, MAE: {mae[i]}")

print(f"Test MAE: {test_mae}, Test RMSE: {np.mean(rmse)}")
```

# transformer model

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.layers import Input, Dense, MultiHeadAttention, Dropout, LayerNormalization, GlobalAveragePooling1D, Flatten
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import MeanSquaredError, MeanAbsoluteError

data_path = 'C:/Users/giris/Downloads/RESEARCH/Personality/personality_data/Video_Interview_features.csv'
data = pd.read_csv(data_path)

X = data.drop(['file_name', 'Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness'], axis=1)
y = data[['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    # Multi-head self-attention
    x = MultiHeadAttention(num_heads=num_heads, key_dim=head_size, dropout=dropout)(inputs, inputs)
    x = Dropout(dropout)(x)
    x = LayerNormalization(epsilon=1e-6)(inputs + x)

    # Feed-forward neural network
    ff_output = Dense(ff_dim, activation="relu")(x)
    ff_output = Dropout(dropout)(ff_output)
    ff_output = Dense(inputs.shape[-1])(ff_output)
    x = LayerNormalization(epsilon=1e-6)(x + ff_output)
    return x

def build_model(input_shape, head_size, num_heads, ff_dim, dropout, mlp_units, output_units):
    inputs = Input(shape=input_shape)
    x = transformer_encoder(inputs, head_size=head_size, num_heads=num_heads, ff_dim=ff_dim, dropout=dropout)
    x = GlobalAveragePooling1D(data_format="channels_first")(x)
    x = Flatten()(x)
    for units in mlp_units:
        x = Dense(units, activation="relu")(x)
        x = Dropout(dropout)(x)
    outputs = Dense(output_units, activation="linear")(x)
    return Model(inputs, outputs)

# Model configuration
input_shape = (X_train.shape[1], 1)
head_size = 64
num_heads = 4
ff_dim = 64
dropout_rate = 0.1
mlp_units = [64, 32]
output_units = y_train.shape[1]

model = build_model(input_shape=input_shape, head_size=head_size, num_heads=num_heads, ff_dim=ff_dim, 
                    dropout=dropout_rate, mlp_units=mlp_units, output_units=output_units)

model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])

model.summary()
X_train_reshaped = np.expand_dims(X_train, -1)
X_test_reshaped = np.expand_dims(X_test, -1)

history = model.fit(X_train_reshaped, y_train, validation_split=0.2, epochs=1, batch_size=128, verbose=1)

test_loss, test_mae = model.evaluate(X_test_reshaped, y_test, verbose=0)
print(f"Test Loss: {test_loss}, Test MAE: {test_mae}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

data_path = 'C:/Users/giris/Downloads/RESEARCH/Personality/personality_data/Extracted_Audio_features.csv'
data = pd.read_csv(data_path)

X = data.drop(['name', 'Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness'], axis=1)
y = data[['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define the MLP model
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(y.shape[1], activation='linear')  # Output layer: one neuron per trait
])

model.compile(optimizer=Adam(learning_rate=0.0001), loss='mse', metrics=['mae'])
history = model.fit(X_train, y_train, validation_split=0.2, epochs=1, batch_size=128, verbose=1)

test_loss, test_mae = model.evaluate(X_test, y_test, verbose=0)

y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred, multioutput='raw_values'))
mae = mean_absolute_error(y_test, y_pred, multioutput='raw_values')

for i, trait in enumerate(['Extraversion', 'Agreeableness', 'Conscientiousness', 'Neuroticism', 'Openness']):
    print(f"{trait} - RMSE: {rmse[i]}, MAE: {mae[i]}")

print(f" Test MAE: {test_mae}, Test RMSE: {np.mean(rmse)}")
```
